```python
# Task 2 Data cleaning

# Deleting columns with missing values in range (70%-100%) for numerical
# values

df = df.drop(columns = [
    "member_id",
    "orig_projected_additional_accrued_interest",
    "hardship_length",
    "hardship_dpd",
    "deferral_term",
    "hardship_amount",
    "hardship_last_payment_amount",
    "hardship_payoff_balance_amount",
    "settlement_term",
    "settlement_amount",
    "settlement_percentage",
    "sec_app_mths_since_last_major_derog",
    "sec_app_revol_util",
    "revol_bal_joint",
    "sec_app_fico_range_low",
    "sec_app_inq_last_6mths",
    "sec_app_mort_acc",
    "sec_app_open_acc",
    "sec_app_fico_range_high",
    "sec_app_open_act_il",
    "sec_app_num_rev_accts",
    "sec_app_chargeoff_within_12_mths",
    "sec_app_collections_12_mths_ex_med",
    "dti_joint",
    "annual_inc_joint",
    "mths_since_last_record",
    "mths_since_recent_bc_dlq",
    "mths_since_last_major_derog",
    "mths_since_recent_revol_delinq"
], axis = 1)
```

```python
# Deleting columns with missing values in range (70%-100%) for categorical
# values
df = df.drop(columns = [
    "hardship_end_date",
    "payment_plan_start_date",
    "hardship_type",
    "hardship_reason",
    "hardship_status",
    "hardship_start_date",
    "hardship_loan_status",
    "settlement_date",
    "debt_settlement_flag_date",
    "settlement_status",
    "sec_app_earliest_cr_line",
    "verification_status_joint",
    "desc"
], axis = 1)
```

```python
num_vars_1 = df.columns[df.dtypes != 'object']
cat_vars_1 = df.columns[df.dtypes == 'object']
```

```python
per_df_1 = df[num_vars_1].isnull().sum().sort_values(ascending=False)/len(df)
per_df_1.head(50)
```

```python
df[cat_vars_1].isnull().sum().sort_values(ascending=False)/len(df)
```

```python
# Converting object date columns to datetime
# Similary with other date columns

df['issue_d'] = pd.to_datetime(df['issue_d'])
df['next_pymnt_d'] = pd.to_datetime(df['next_pymnt_d'])
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'])
df['last_credit_pull_d'] = pd.to_datetime(df['last_credit_pull_d'])
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'])
df['issue_d'] = pd.to_datetime(df['issue_d'])
```

```python
df[cat_vars_1]
```

```python
# Replacing null values to median() for numerical values

for column in df[num_vars_1]:
  median_num_vars = df[column].median()
  df[column].fillna(median_num_vars, inplace=True)
```

```python
# Checking replace for numerical values

df[num_vars_1].isnull().sum().sort_values(ascending=False)/len(df)
```

```python
# Replacing null values to missing_data for catigorical values

df[cat_vars_1] = df[cat_vars_1].fillna("Missing_Data")
```

```python
# Checking replace for categorical values

df[cat_vars_1].isnull().sum().sort_values(ascending=False)/len(df)
```

```python
# There are not duplicate values in dataset

df.duplicated().sum()
```

```python
# Task 3 Data manipulation

# Convert all string data to lowercase

df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# Create a new column loan_grade based on loan amount quantiles

quantiles = df['loan_amnt'].quantile([0, 1/3, 2/3, 1])
df['loan_grade'] = pd.cut(df['loan_amnt'], bins = quantiles, labels = ['low', 'medium', 'high'])

df.head()
```

```python
# Task 4 Data Queries

# Select the rows where the interest rate is lower than AVG for all rates

mean_int_rate = df['int_rate'].mean()
rows = df[df['int_rate'] < mean_int_rate]

# Finding the most frequently state among the loan record

freq_state = df["addr_state"].mode()[0]

print(rows)
print("\n", freq_state)
```

```python
# Task 5 Grouping and aggregation

# Group the data by grade and find AVG loan_amnt and int_rate for each group

group_by_grade = df.groupby('grade').agg({'loan_amnt': 'mean', 'int_rate': 'mean'}).reset_index()
group_by_grade.columns = ['grade', 'avg_loan_amnt', 'avg_int_rate']

# Calculate the total number of loans issued each year

df['issue_d'] = df['issue_d'].dt.year

total_loan_by_year = df.groupby('issue_d').size()

print(group_by_grade)
print("\n", total_loan_by_year)
```

```python
# Task 6 Data Visualization

# Build histogram of loan_amnt

plt.figure(figsize=(10, 6))
sns.histplot(data = df, x='loan_amnt', bins=30, color='blue', edgecolor = 'red')
plt.title('Histogram of loan_amnt')
plt.xlabel('loan_amnt')
plt.ylabel('frequency')
plt.grid(True)
plt.show()

# Create a scatter plot to explore the relationalship between loan_amnt and
# int_rate

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='loan_amnt', y='int_rate', color='orange', alpha=0.8)
plt.title('Relationalship between loan_amnt and int_rate')
plt.xlabel('loan_amnt')
plt.ylabel('int_rate')
plt.grid(True)
plt.show()
```

```python
# Task 7 Advanced analysis

# Some other cleaning and changing data type with columns emp_length and
# loan_status

# Replace

df['emp_length'] = df['emp_length'].replace({
    '2 years':'2',
    '10+ years':'10',
    '1 year':'1',
    '9 years':'9',
    '< 1 year':'0.5',
    '5 years':'5',
    '6 years':'6',
    '3 years':'3',
    '8 years':'8',
    'missing_data':'0',
    '4 years':'4',
    '7 years':'7'
}, regex=True)

# Categorical to numeric

df['emp_length'] = pd.to_numeric(df['emp_length'], errors = 'coerce')

# Mapping values of column loan_status to numerical 0 means low risk 1 means
# high risk

status_map = {
    'charged off': 1,
    'fully paid': 0,
    'late (31-120 days)': 1,
    'current': 0,
    'in grace period': 0,
    'late (16-30 days)': 0,
    'does not meet the credit policy. status:fully paid': 0,
    'does not meet the credit policy. status:charged off': 1,
    'default': 1
}

df['loan_status_encode'] = df['loan_status'].map(status_map)

# Calculate and build a correlation matrix between emp_length and loan_status
corr_matrix = df[['emp_length', 'loan_status_encode']].corr()

# Use boxplot to show the distribution of loan_amnt by loan term

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='term', y='loan_amnt', palette='Set2', hue='term', legend=False)
plt.title('Distribution of loan_amnt by loan term')
plt.xlabel('term')
plt.ylabel('loan_amnt')
plt.grid(True)
plt.show()

print(corr_matrix)
```
